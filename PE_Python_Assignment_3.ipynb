1.	Read mobile dataset using pandas, and preprocess it (fill null values).

```
import pandas as pd
d = pd.read_csv('/content/drive/My Drive/mobileprice.csv')
d.describe()
```

```
  d.isnull()
```

```
  d.isnull().sum()
```

```
  d.isnull().sum().sum()  #total null values
```

```
d['weight']=d['weight'].fillna(d['weight'].mean())
d['stand_by_time']=d['stand_by_time'].fillna(d['stand_by_time'].mean())
d['no_of_reviews_in_gsmarena_in_week']=d['no_of_reviews_in_gsmarena_in_week'].fillna(d['no_of_reviews_in_gsmarena_in_week'].mean())
d['phone_height']=d['phone_height'].fillna(d['phone_height'].mean())
d['talk_time']=d['talk_time'].fillna(d['talk_time'].mean())
d['video_resolution']=d['video_resolution'].fillna(d['video_resolution'].mean())
d['primary_camera_resolution']=d['primary_camera_resolution'].fillna(d['primary_camera_resolution'].mean())
```

```
d.isnull().sum().sum()  #all null values are replaced with mean values
```

2.	Draw correlation graph for mobile price with all attributes.**bold text**

```
import matplotlib.pyplot as plt
p=list(d["price"])
col=list(d.columns)
for n in col :
   c=list(d[n])
   plt.xlabel(n)
   plt.ylabel("price")
   plt.plot(c,p,label=n)
   plt.scatter(c,p)
   plt.legend()
   plt.show()
```

4.	Write a NumPy program to compute the condition number of a given matrix.

```
matrix = np.array([[1, 2], [3, 4]])
print("Original matrix: ")
print(matrix) 
print("Condition number of the saidmatrix: ")
finalres = np.linalg.cond(matrix)
print(finalres)
```
